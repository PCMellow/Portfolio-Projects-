# Automating Crypto Website API Pull Using Python

```python
from requests import Request, Session
from requests.exceptions import ConnectionError, Timeout, TooManyRedirects
import json

url = 'https://pro-api.coinmarketcap.com/v1/cryptocurrency/listings/latest'
parameters = {
  'start':'1',
  'limit':'10',
  'convert':'USD'
}
headers = {
  'Accepts': 'application/json',
  'X-CMC_PRO_API_KEY': '3a22c3df-e63d-43a3-958f-244b97926599',
}

session = Session()
session.headers.update(headers)

try:
  response = session.get(url, params=parameters)
  data = json.loads(response.text)
  print(data)
except (ConnectionError, Timeout, TooManyRedirects) as e:
  print(e)
```

```python
def api_runner():
    global dataframe
    url = 'https://pro-api.coinmarketcap.com/v1/cryptocurrency/listings/latest'
    parameters = {
      'start':'1',
      'limit':'10',
      'convert':'USD'
    }
    headers = {
      'Accepts': 'application/json',
      'X-CMC_PRO_API_KEY': '3a22c3df-e63d-43a3-958f-244b97926599',
    }

    session = Session()
    session.headers.update(headers)

    try:
      response = session.get(url, params=parameters)
      data = json.loads(response.text)
      print(data)
    except (ConnectionError, Timeout, TooManyRedirects) as e:
      print(e)
    
    dataframe2 = pd.json_normalize(data['data'])
    dataframe2['Timestamp'] = pd.to_datetime("now",utc=True)
    dataframe = pd.concat([dataframe, dataframe2])
    
    # first checking for the csv file, if not then create, if already present then append the data in the same file
    if not os.path.isfile(r'C:\Users\Pankaj Chettri\Desktop\STUDY MATERIALS\Knowledge\Work Materials\PYTHON\Jupyter\API.csv'):
        dataframe.to_csv(r'C:\Users\Pankaj Chettri\Desktop\STUDY MATERIALS\Knowledge\Work Materials\PYTHON\Jupyter\API.csv', header = 'column_names')
    else:
        dataframe.to_csv(r'C:\Users\Pankaj Chettri\Desktop\STUDY MATERIALS\Knowledge\Work Materials\PYTHON\Jupyter\API.csv',mode = 'a', header = False)
```
